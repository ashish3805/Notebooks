```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.xlabel('some x')
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25])
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'r-')
plt.axis([0,6,0,20])
plt.show()
```

```python
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
